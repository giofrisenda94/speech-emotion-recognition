```python
X_rms_feature_list = np.array(new_df["zero_crossing_rate"])
y_rms_feature_list = new_df["emotions"]
X_rms_feature_list[0].shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_rms_feature_list, y_rms_feature_list, test_size = 0.3, random_state = 1)

X_train = list(X_train)

X_train = np.transpose(np.array(X_train), (0, 2, 1))
X_train.shape
```

```python
X_test = list(X_test)

X_test = np.transpose(np.array(X_test), (0, 2, 1))
X_test.shape
```

```python
#'One-hot' vectors for Y: emotion classification
y_train_class = to_categorical(y_train, dtype = 'int8')
y_test_class = to_categorical(y_test, dtype = 'int8')

```

```python
es = EarlyStopping(patience=20, restore_best_weights=True)

model = Sequential()
   
model.add(Conv1D(256, 5,padding='same',
                     input_shape=(448, 1)))
model.add(Activation('relu'))
model.add(Conv1D(128, 5,padding='same'))
model.add(Activation('relu'))
model.add(Dropout(0.1))
model.add(MaxPooling1D(pool_size=(8)))
model.add(Conv1D(128, 5,padding='same',))
model.add(Activation('relu'))
model.add(Conv1D(128, 5,padding='same',))
model.add(Activation('relu'))
model.add(Conv1D(128, 5,padding='same',))
model.add(Activation('relu'))
model.add(Dropout(0.2))
model.add(Conv1D(128, 5,padding='same',))
model.add(Activation('relu'))
model.add(Flatten())
model.add(Dense(9))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.fit(X_train, y_train_class,
          epochs=300,
          batch_size=128,
          validation_split=0.3,
          verbose=1,
          callbacks=[es]
         )
    
```
